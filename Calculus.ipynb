The question for the quiz is the integration of indifinite integral of function e ^ x * sin (x)

It's been such long time since i work on solving calculus question so naturally I don't remember how to solve them and I have to look up how to do it.
You can solve it by using integration by parts. 


```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = Symbol('x')
integrate(exp(x) * sin(x), x)
```

The good thing about ipython is not only that you can solve standard calculus problem quite quickly.
You can also plot quite easily common functions and that helps for visualization purpose.

Let's see how each of the curve looks like when we plot them over the range from -2pi to 2pi

```python
x = linspace(-2*np.pi, 2*np.pi, 150)
```

```python
plt.plot(x, np.exp(x))
```

Sin curve is of course sinusoidal. 

```python
plt.plot(x, np.sin(x))
```

Let's now multiply the two curves together.
I've initially thought that the peak of the sin curve (magnitude of 1.0 or -1.0) will correspond to the peak of the curve before changing direction
But this is not the case.

```python
plt.plot(x, np.exp(x)*np.sin(x))
plt.xlim(-2*np.pi, 0)
plt.ylim(-0.35, 0.025)
```

```python
plt.plot(x, np.exp(x)*np.sin(x))
plt.xlim(0, 2*np.pi)
plt.ylim(-200, 20)
```

As we extend X-axis further, the Y-axis grow at exponential rate. It still behaves like sinusoidal curve although much harder to spot.


```python
plt.plot(x, np.exp(x)*np.sin(x))
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-200, 20)
```
